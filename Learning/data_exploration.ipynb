You can select multiple "features"/values to reduce the amount of noise in our data models. you can do this by making an array of the desired features. By convention the array should be named "X"

```python
house_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = house_data[house_features]
X.describe()
```

A library called scikit-learn can be used to create the models based off of the data given in our dataframes. To use the  library use this code snippet: we first import the library. Then we create the variable for our model which we will use on our data. Then we fit the model with our target prediction and our features.

```python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
house_model = DecisionTreeRegressor(random_state=1)

# Fit model
house_model.fit(X, y)
```

Once we have specified what we want to predict, y (in this case ), and our features used to predict the target prediction, X we can call the predict method on the dataset with X as a parameter.

```python
house_model.predict(X.head())
```

There are many ways to check the accuracy/quality of a model. One of the simplist ways of doing this is using the MAE metric short for Mean Absolute Error. Simply put MAE can be calculated by subtracting the actual value and your predicted value. e.g. A house is predicted to have a value of 250,000 while it actually has a value of 300,000. this would give you a MAE metric of 50,000. We repeat that for all predictions made for the data frame and average it out.

We can use the sklearn library again to find the MAE metric of a model very easily.

```python
from sklearn.metrics import mean_absolute_error

predicted_home_prices = house_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

It is good practice to split your training data into two pieces. One half should be used as training data while the other half should be used as validating data. This is because it will prevent the model from having strange biases about factors that may truely not have an affect on the target prediction. Once again this can be implemented with the sklearn library.

```python
from sklearn.model_selection import train_test_split

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
# Define model
house_model = DecisionTreeRegressor()
# Fit model
house_model.fit(train_X, train_y)

# get predicted prices on validation data
val_predictions = house_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

This shows the problems with using a very simple metric to measure the accuracy of your model. With the MAE metric the predictions were only off by around ~1000 which is extremely accurate (within > %1>) while if we tried using out of sample data our estimates were much further off (250,000! almost a quarter the cost of a house).

These trees are fairly simply and have just been spliting into two parts at each branch. This is an okay way to make simple predictions however as the tree gets more and more branches we lead to a problem called overfitting. Overfitting simply means that we are getting too precise with how we are spliting our data thereby limiting the amount of data we have on specific conditions for each leaf of the decision tree.

Additionally we can also run into a problem called underfitting at the other end of the spectrum where we have so little division that there is no real correlation between anything in our data groups to allow for acurate predictions.

A simple way to control the accuracy to prevent us from overfitting or underfitting our data is to use the max_lead_nodes to prevent us from creating too many potential data groups.

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

Once we find the optimal number of leaves in our decision tree we can now use all of our data (yes including our validation data) in the final deployment of our decision tree. When we use our decision tree regressor function make sure to use the max_leaf_nodes argument and set it to the optimal amount of leaves to get the best accuracy

```python
# Fill in argument to make optimal size and uncomment
final_model = DecisionTreeRegressor(max_leaf_nodes=100, random_state=0)

# fit the final model and uncomment the next two lines
final_model.fit(X, y)
```
