This is the basic setup for using pandas. First you import of course. Then you set the path location of your data. Next you read the data from the file using Pandas then use the describe method to print the data

```python
import pandas as pd

house_data_location = "./melb_data.csv"

house_data = pd.read_csv(house_data_location)

house_data.describe()
```

using the columns property allows you to find the entries on the top of the charts


```python
house_data.columns
```

using dropna(aix=*axis to drop*) will remove missing values from the data

```python
house_data.dropna(axis=1)
```

To select the column you want to predict using the data use dot notation. This is called the target prediction and should be saved in the variable "y" by convention.

```python
y = house_data.Price
y
```

You can select multiple "features"/values to reduce the amount of noise in our data models. you can do this by making an array of the desired features. By convention the array should be named "X"

```python
house_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = house_data[house_features]
X.describe()
```

You can use the head() function on X or the "features" in order to show a sample of the first few pieces of data in the dataframe.

```python
X.head()
```

A library called scikit-learn can be used to create the models based off of the data given in our dataframes. To use the  library use this code snippet: we first import the library. Then we create the variable for our model which we will use on our data. Then we fit the model with our target prediction and our features.

```python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
house_model = DecisionTreeRegressor(random_state=1)

# Fit model
house_model.fit(X, y)
```

Once we have specified what we want to predict, y (in this case ), and our features used to predict the target prediction, X we can call the call the predict method on the dataset with X as a parameter.

```python
house_model.predict(X.head())
```
